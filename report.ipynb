```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Dataframe PPO

```python
attr = ['time/total_timesteps','time/time_elapsed','rollout/ep_rew_mean']

ppo_0 = pd.read_csv('./results/ppo/0/progress.csv')
ppo_0 = ppo_0[attr]
ppo_0['agent'] = 'PPO'
ppo_1 = pd.read_csv('./results/ppo/1/progress.csv')
ppo_1 = ppo_1[attr]
ppo_1['agent'] = 'PPO'
ppo_2 = pd.read_csv('./results/ppo/2/progress.csv')
ppo_2 = ppo_2[attr]
ppo_2['agent'] = 'PPO'
ppo_3 = pd.read_csv('./results/ppo/3/progress.csv')
ppo_3 = ppo_3[attr]
ppo_3['agent'] = 'PPO'
ppo_4 = pd.read_csv('./results/ppo/4/progress.csv')
ppo_4 = ppo_4[attr]
ppo_4['agent'] = 'PPO'
```

# Dataframe DQN

```python
attr = ['time/total_timesteps','time/time_elapsed','rollout/ep_rew_mean']

dqn_0 = pd.read_csv('./results/dqn/0/progress.csv')
dqn_0 = dqn_0[attr]
dqn_0['agent'] = 'DQN'
dqn_1 = pd.read_csv('./results/dqn/1/progress.csv')
dqn_1 = dqn_1[attr]
dqn_1['agent'] = 'DQN'
dqn_2 = pd.read_csv('./results/dqn/2/progress.csv')
dqn_2 = dqn_2[attr]
dqn_2['agent'] = 'DQN'
dqn_3 = pd.read_csv('./results/dqn/3/progress.csv')
dqn_3 = dqn_3[attr]
dqn_3['agent'] = 'DQN'
dqn_4 = pd.read_csv('./results/dqn/4/progress.csv')
dqn_4 = dqn_4[attr]
dqn_4['agent'] = 'DQN'
```

## Concatenando os dataframes

```python
df = pd.concat([
    ppo_0,
    ppo_1,
    ppo_2,
    ppo_3,
    ppo_4,
    dqn_0,
    dqn_1,
    dqn_2,
    dqn_3,
    dqn_4
])

df
```

## Plot DQN x PPO

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x="time/total_timesteps", y="rollout/ep_rew_mean", hue='agent', data=df)
plt.xlabel('Timesteps')
plt.ylabel('Recompensa')
plt.title('Comparando DQN e PPO no ambiente Minigrid DoorKey-6x6-v0')
plt.legend(title='Algoritmo')
plt.savefig('docs/assets/plot.png')
plt.show()
```
